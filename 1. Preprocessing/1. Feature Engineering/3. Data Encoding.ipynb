### Data Encoding
In this project, categorical and binary variables in the dataset are encoded using numerical values to make them suitable for machine learning models. The encoding scheme follows a label encoding approach, where categorical variables are converted into numeric representations.

### Encoding Methods Used
- Label Encoding: Categorical variables (e.g., gender, class) are assigned numeric values.

- Normalization (Optional): If required, features like age can be normalized using Min-Max Scaling or Standardization to improve model performance.



```python
## Importing Basic Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
## Loading Dataset
dataset = pd.read_csv("/home/ghost/Documents/Data Mining Mini Project/data/diabetes_data.csv")
```

```python
dataset.head()
```

```python
# Apply Label Encoding to binary categorical features
binary_features = ["Gender", "Polyuria","Polydipsia","sudden weight loss", 
                   "weakness", "Polyphagia", "Genital thrush", 
                   "visual blurring","Itching", "Irritability", 
                   "delayed healing", "partial paresis", 
                   "muscle stiffness","Alopecia","Obesity","class"]
```

```python
### Initializing LabelEncoder
encoder = LabelEncoder()
```

```python
## Applying LabelEncoder
for item in binary_features:
    dataset[item] = encoder.fit_transform(dataset[item])
```

```python
## Lets Save the Clean Dataset
dataset.to_csv("/home/ghost/Documents/Data Mining Mini Project/data/diabetes_data_encoded.csv", index=False)
```
